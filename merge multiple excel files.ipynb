```python
import pandas as pd
import glob
#return all files in "sheets" folder using *
file_list= glob.glob(r"C:\Users\HP\Desktop\sheets/*")
file_list
```

```python
#combine all files
files = []
for filename in file_list:
    df = pd.read_excel(filename)
    files.append(df)
frame = pd.concat(files, axis=0, ignore_index=True)
```

```python
frame
```

```python
#save frame as an excel file
```

```python
frame.to_excel(r"C:\Users\HP\Desktop\frame.xlsx",index =False)
```
